```python
import pandas as pd
```

```python
df = pd.read_csv("top_steam_games_fixed.csv")
```

```python
df.head()
```

```python
df = df[df["Game Name"].notna()]
```

```python
df.head()
```

```python
df["Game Tags"] = df["Game Tags"].astype(str)
df["Game Tags"] = df["Game Tags"].str.replace(r"[\t\n\r]+", " ", regex=True)
df["Game Tags"] = df["Game Tags"].str.replace(r"\+", "", regex=True)
df["Game Tags"] = df["Game Tags"].str.replace(r"\s{2,}", " ", regex=True)
df["Game Tags"] = df["Game Tags"].str.strip()
```

```python
df["Game Name"] = df["Game Name"].str.replace("â„¢", "™", regex=False)
df["Game Name"] = df["Game Name"].str.replace("Â®", "®", regex=False)

df["Price"] = df["Price"].astype(str).str.replace("â‚¹", "₹", regex=False)
df["Price"] = df["Price"].str.strip()
```

```python
df = df.drop(columns=["Recent Reviews", "All Reviews"], errors='ignore')

df = df.rename(columns={"Price": "Price (in Rupees)"})
```

```python
df = df.reset_index(drop=True)
df.to_csv("steam_games.csv", index=False)

print("Cleaned file saved as 'steam_games.csv'")
```
